## Spline crash course

First thing to do is to transform the space into B-spline basis functions... $b_k(x)$

```python
%matplotlib inline
import matplotlib.pyplot as plt
from concise.utils.splines import BSpline
from concise.preprocessing import encodeSplines # uses BSpline
```

```python
def plot_order(degree, ax):
    x = np.linspace(0, 1, 100)
    y = encodeSplines(x, 10, degree).squeeze() # alternatively: y = BSpline(0, 1, 10, degree).predict(x)
    ax.plot(x, y)
    ax.set(title="Degree: {d}, order: {o}".format(d=degree, o=degree+1))
```

```python
fig, ax = plt.subplots(nrows=4)
fig.tight_layout()
for i in range(ax.shape[0]):
    plot_order(i, ax[i])
```
